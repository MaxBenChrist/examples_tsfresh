```python
import tsfresh
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
df = pd.DataFrame(np.concatenate([np.repeat(range(500),40)[:,np.newaxis],
                                  np.tile(range(40),500)[:,np.newaxis],
                                  np.random.randn(20000, 1)], axis=1), columns=["id", "time", "x"])

df
```

```python
ts = tsfresh.extract_features(df, column_id="id")
# ts_filtered = tsfresh.extract_relevant_features(df, y, column_id="id")
```

```python
df.to_csv("df.csv", index=False)
ts.to_csv("ts.csv", index=False)
# ts_filtered.to_csv("ts_filtered.csv", index=False)
```
